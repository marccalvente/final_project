```python
import pandas as pd
import numpy as np
import matplotlib as plt

import seaborn as sns
sns.set_theme()
```

```python
positions = pd.read_csv("./data/position_tracking.csv", index_col="Unnamed: 0", converters={'0': pd.eval})
```

```python
positions["x"] = positions["0"].apply(lambda coords: coords[0])
positions["y"] = positions["0"].apply(lambda coords: coords[1])
positions.drop("0", axis=1, inplace=True)
```

```python
ax = sns.scatterplot(data=positions, x="x", y="y", color="gold")
ax.set(title='Tracked positions')
ax.set(xlabel="X position", ylabel="Y position", xlim=(0, 640), ylim=(0, 480))
ax.set()
fig = ax.get_figure()
fig.savefig('./images/tracked_positions_plot.png',bbox_inches='tight');
```

```python
positions.to_csv("./data/position_tracking_clean.csv")
```
